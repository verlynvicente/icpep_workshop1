```python
# Normalize data (length of 1) 
import matplotlib.pyplot as plt 
import numpy
from sklearn.preprocessing import Normalizer 
from pandas import read_csv 
from numpy import set_printoptions 
filename = 'Student_Admissions.csv'   
dataframe = read_csv(filename) 

#dropping of columns
df=dataframe.drop(columns=['0','Batch','UP'])
#dropping
dataframe2=df.dropna(how='any')

#check if there's null
dataframe2.isnull().sum()
#print(dataframe2)

array = dataframe2.values 

# separate array into input and output components 
X = array[:,0:13] 
Y = array[:,13] 
scaler = Normalizer().fit(X) 
normalizedX = scaler.transform(X) 

# summarize transformed data 
set_printoptions(precision=3) 
normalizedX[0:5,:]

myarray = normalizedX[0:5,:] 
plt.plot(myarray) 
plt.xlabel(' x axis' ) 
plt.ylabel(' y axis' ) 
plt.show()
```

```python
# Feature Extraction with Univariate Statistical Tests (Chi-squared for classification) 
from pandas import read_csv 
from numpy import set_printoptions 
from sklearn.feature_selection import SelectKBest 
from sklearn.feature_selection import chi2 

#filename = 'Student_Admissions.csv'
#dataframe = read_csv(filename) 
array = dataframe2.values 
X = array[:,0:13] 
Y = array[:,13] 

# feature extraction 
test = SelectKBest(score_func=chi2, k=4) 
fit = test.fit(normalizedX, Y) 

# summarize scores
set_printoptions(precision=3) 
print(fit.scores_) 
features = fit.transform(normalizedX) 

# summarize selected features 
print(features[0:5,:])
pd.DataFrame(fit.scores_.reshape(1, 13).tolist(), columns=dataframe2.columns[:-1]).T.sort_values(0, ascending=False)
```

```python
# Evaluate using a train and a test set 
from pandas import read_csv 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
#filename = 'pima-indians-diabetes.data.csv' 
#names = ['preg' , 'plas' , 'pres' , 'skin' , 'test' , 'mass' , 'pedi' , 'age' , 'class'] 
#dataframe = read_csv(filename, names=names) 
array = dataframe2.values 
X = array[:,0:13] 
Y = array[:,13] 
test_size = 0.33 
seed = 7 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
model = LogisticRegression() 
model.fit(X_train, Y_train) 
result = model.score(X_test, Y_test) 
print("Accuracy: %.3f%%" % (result*100.0))
```

```python
# Evaluate using Cross Validation 
from pandas import read_csv 
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LogisticRegression 
#filename = 'pima-indians-diabetes.data.csv' 
#names = ['preg' , 'plas' , 'pres' , 'skin' , 'test' , 'mass' , 'pedi' , 'age' , 'class'] 
#dataframe = read_csv(filename, names=names) 
array = dataframe2.values 
X = array[:,0:13] 
Y = array[:,13] 
kfold = KFold(n_splits=10, random_state=7) 
model = LogisticRegression() 
results = cross_val_score(model, X, Y, cv=kfold) 
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
# Compare Algorithms 
from pandas import read_csv 
from matplotlib import pyplot 
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
from sklearn.naive_bayes import GaussianNB 
from sklearn.svm import SVC 

# load dataset 
#filename = 'pima-indians-diabetes.data.csv' 
#names = ['preg' , 'plas' , 'pres' , 'skin' , 'test' , 'mass' , 'pedi' , 'age' , 'class'] 
#dataframe = read_csv(filename, names=names) 
array = dataframe2.values 
X = array[:,0:13] 
Y = array[:,13] 

# prepare models 
models = [] 
models.append((' LR' , LogisticRegression())) 
models.append((' LDA' , LinearDiscriminantAnalysis())) 
models.append((' KNN' , KNeighborsClassifier())) 
models.append((' CART' , DecisionTreeClassifier())) 
models.append((' NB' , GaussianNB())) 
models.append((' SVM' , SVC()))

# evaluate each model in turn 
results = [] 
names = [] 
scoring = 'accuracy' 
for name, model in models: 
    kfold = KFold(n_splits=12, random_state=10) 
    cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring) 
    results.append(cv_results)
    names.append(name) 
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std()) 
    print(msg)
# boxplot algorithm comparison 
fig = pyplot.figure() 
fig.suptitle('Algorithm Comparison') 
ax = fig.add_subplot(111) 
pyplot.boxplot(results) 
ax.set_xticklabels(names)
pyplot.show()
```
